```python
from pickle import load
from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
data = load(open('../Data/SouthPark_Data.pkl', 'rb'))
```

```python
tokenized = tokenizer.tokenize(data[list(data.keys())[0]][3])
```

```python
text = ""
for token in tokenized:
    text+=token+" "

print(text)
```
